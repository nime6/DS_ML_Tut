```python
"""
# Basic Data Manipulation and Visualization

This Jupyter Notebook introduces fundamental data science techniques using Pandas and Matplotlib, using an engineering dataset.
"""

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Define color scheme
colors = ["#DCF6F6", "#C5E5FC", "#0A75C4", "#0025AA"]

```

```python

"""
## Engineering Data Example
We will use a dataset representing **material stress vs. strain** values, which are crucial in mechanical engineering.
"""

# Generate sample stress-strain dataset
strain = np.linspace(0, 0.02, 10)  # Strain in percentage
stress = 200000 * strain  # Stress in Pascals (Hookeâ€™s Law for linear elasticity)

df = pd.DataFrame({
    'Strain (%)': strain * 100,  # Convert to percentage
    'Stress (Pa)': stress
})

display(df.head())
```

```python

"""
## Data Manipulation
### Sorting Data
Sort the dataset by stress values in descending order.
"""
df_sorted = df.sort_values(by='Stress (Pa)', ascending=False)
display(df_sorted)
```

```python

"""
### Filtering Data
Filter the dataset to include only stress values above a certain threshold.
"""
threshold = 2000  # Change this value to filter more or fewer rows
df_filtered = df[df['Stress (Pa)'] > threshold]
display(df_filtered)
```

```python

"""
## Data Visualization
### Stress-Strain Curve (Line Plot)
A fundamental plot in mechanical engineering, showing material deformation.
"""
plt.figure(figsize=(6,4))
plt.plot(df['Strain (%)'], df['Stress (Pa)'], marker='o', linestyle='-', color=colors[2])
plt.xlabel('Strain (%)')
plt.ylabel('Stress (Pa)')
plt.title('Stress-Strain Curve')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python

"""
### Histogram of Stress Values
Understanding the distribution of stress values.
"""
plt.figure(figsize=(6,4))
plt.hist(df['Stress (Pa)'], bins=5, color=colors[1], edgecolor=colors[3])  # Change bins for more/less granularity
plt.xlabel('Stress (Pa)')
plt.ylabel('Frequency')
plt.title('Stress Value Distribution')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
